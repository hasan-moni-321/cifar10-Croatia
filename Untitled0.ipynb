```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot
import matplotlib.pyplot as plt
from PIL import Image

import collections 
%matplotlib inline 

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix

import keras
import tensorflow 
from tensorflow.keras.datasets import cifar10
# from keras.utils import to_categorical
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

from keras.layers import Activation, Dense

from keras.models import Sequential, Input, Model
from keras.layers import Dense, Dropout, Flatten, Lambda
from keras.layers import Conv2D, MaxPooling2D
# from keras.layers.normalization import BatchNormalization
from tensorflow.keras.layers import BatchNormalization
from keras.layers.advanced_activations import LeakyReLU
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.optimizers import SGD
from keras.utils.vis_utils import plot_model
from tensorflow.keras.models import load_model 
from tensorflow.keras.preprocessing.image import ImageDataGenerator 
from tensorflow.keras.applications import DenseNet121, DenseNet201, VGG19
from keras.initializers import he_normal
from keras.regularizers import l2
from tensorflow.keras.optimizers import Adam 
#import tensorflow as tf
from tensorflow.keras.applications.vgg19 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image 
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.models import load_model

#from tensorflow.keras.losses import CategoricalCrossentropyLoss
```

### 1. Loading The Dataset

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data() 
print("Shape of x_train is : {}".format(x_train.shape)) 
print("Shape of x_test is : {}".format(x_test.shape)) 
print("Shape of y_train is : {}".format(y_train.shape)) 
print("Shape of y_test is : {}".format(y_test.shape)) 
```

### 2. Unique Classes

### 4. Converting to float32

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32') 
print("Shape of x_train is :{}".format(x_train.shape)) 
print("Shape of x_test is :{}".format(x_test.shape)) 
```

### 5. Rescale Pixel Value in range 0-1

```python
x_train = x_train / 255
x_test = x_test / 255

print("Shape after rescaling :{}".format(x_train.shape))
print("Shape after rescaling :{}".format(x_test.shape))
```

### 6. Spliting train data into train and validation data

```python
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=.1, random_state=13)  
print("Shape of x_train is :{}".format(x_train.shape)) 
print("Shape of x_valid is :{}".format(x_valid.shape)) 
print("Shape of y_train is :{}".format(y_train.shape)) 
print("Shape of y_valid is :{}".format(y_valid.shape)) 
```

### 7. categorical to one-hot encoding

```python
# flatten the label values
# y_train, y_test = y_train.flatten(), y_test.flatten()
# print(y_train.shape, y_test.shape) 


y_train = to_categorical(y_train)
y_valid = to_categorical(y_valid)
y_test = to_categorical(y_test) 
```

### 8. Building Models 

```python
i = Input(shape=x_train[0].shape)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(i)
x = BatchNormalization()(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = BatchNormalization()(x)
x = MaxPooling2D((2, 2))(x)

x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = BatchNormalization()(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = BatchNormalization()(x)
x = MaxPooling2D((2, 2))(x)
 
x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
x = BatchNormalization()(x)
x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
x = BatchNormalization()(x)
x = MaxPooling2D((2, 2))(x)
x = Flatten()(x)
x = Dropout(0.2)(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.2)(x)
x = Dense(10, activation='softmax')(x)
 
model = Model(i, x)
 
```

### 9. Summary of the Model 

```python
# model summary
model.summary()
```

### 10. Plot Model

```python
#plotting model
plot_model(model, show_shapes=True, to_file = 'model_1_plot_cifar10.png')
```

### 11. Compile Model

```python
# Compile model 
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

### 12. Train Model 

```python
batch_size = 32
data_generator = ImageDataGenerator(
  width_shift_range=0.1, height_shift_range=0.1, horizontal_flip=True)
 
train_generator = data_generator.flow(x_train, y_train, batch_size)
steps_per_epoch = x_train.shape[0] // batch_size
 
model_history = model.fit(train_generator, validation_data=(x_valid, y_valid),
              steps_per_epoch=steps_per_epoch, epochs=20)
```

### 13. Training Accuracy and Loss Graph

```python
# Accuracy and Loss graph For Model_1 
accuracy = model_history.history['accuracy']
val_accuracy = model_history.history['val_accuracy']

loss = model_history.history['loss']
val_loss = model_history.history['val_loss']

epochs = range(len(accuracy))
plt.plot(epochs, accuracy, 'go', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'g', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'go', label='Training loss')
plt.plot(epochs, val_loss, 'g', label='validation loss')
plt.title('training and validation loss')
plt.legend()
plt.show()
```

### 14. Evaluate with Test Data

```python
# For model_1 
model_test_eval = model.evaluate(x_test, y_test, verbose=1)
print("Test loss is :{}".format(model_test_eval[0])) 
print("Test accuracy is :{}".format(model_test_eval[1]))   
```

# Using Pre-trained Model()

### 1. Loading Dataset

```python
# loading dataset 
(x_train,y_train), (x_test,y_test)=cifar10.load_data() 
```

### 2. Dividing Dataset

```python
# Dividing Dataset 
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,test_size=.1)
print("Shape of x_train is :{} \nShape of x_valid is :{} \nShape of x_test is :{} \nShape of y_train is :{} \nShape of y_valid is :{} \nShape of y_test is :{}".format(x_train.shape , x_val.shape, x_test.shape, y_train.shape, y_val.shape, y_test.shape)) 
```

### 3. Encoding

```python
y_train=to_categorical(y_train)
y_val=to_categorical(y_val)
y_test=to_categorical(y_test)
```

### 4. Define Generator

```python
# Train Generator 
train_generator = ImageDataGenerator(
                                    rotation_range=2, 
                                    horizontal_flip=True,
                                    zoom_range=.1 )
train_generator.fit(x_train)

# valid Generator 
val_generator = ImageDataGenerator(
                                    rotation_range=2, 
                                    horizontal_flip=True,
                                    zoom_range=.1)
val_generator.fit(x_val)

# Test Generator 
test_generator = ImageDataGenerator(
                                    rotation_range=2, 
                                    horizontal_flip= True,
                                    zoom_range=.1) 
test_generator.fit(x_test)
```

### 5. Define Model

```python
# Define vgg19 Model 
base_model = VGG19(include_top=False,
                     weights='imagenet',
                     input_shape=(32,32,3),
                     classes=y_train.shape[1])


model_pretrained = Sequential()
model_pretrained.add(base_model) 
model_pretrained.add(Flatten()) 

model_pretrained.add(Dense(1024,activation=('relu'),input_dim=512))
model_pretrained.add(Dense(512,activation=('relu'))) 
model_pretrained.add(Dense(256,activation=('relu'))) 
#model.add(Dropout(.2)) 
model_pretrained.add(Dense(128,activation=('relu')))
#model.add(Dropout(.2))
model_pretrained.add(Dense(10,activation=('softmax'))) 
```

### 6. Model Summary

```python
# model summary 
model_pretrained.summary() 
```

### 7. Plotting Model

```python
#plotting model
plot_model(model_pretrained, show_shapes=True, to_file = 'plot_cifar10_pretrained.png')
```

### 8. compile Model

```python
learn_rate=.001

sgd=SGD(learning_rate=learn_rate,momentum=.9,nesterov=False)
adam=Adam(learning_rate=learn_rate, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)

# compile the model 
model_pretrained.compile(optimizer=sgd,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

### 9. Define Callbacks

```python
# define ReduceLROnPlateau, EarlyStopping, ModelCheckpoint 
lr_reduce = ReduceLROnPlateau(monitor='val_loss', 
                       factor=.01,
                       patience=3,  
                       min_lr=1e-5)

# early_stop = EarlyStopping(monitor='val_accuracy',
#                                        patience=3, 
#                                        verbose=1,
#                                        mode='max')

# checkpoint = ModelCheckpoint('cifar10.h5',
#                                          monitor='val_accuracy',
#                                          verbose=1,
#                                          save_weights_only=False,
#                                          save_best_only=True,
#                                          mode='max',
#                                          save_freq='epoch') 
```

### 10. Training The Model

```python
# some variables
batch_size= 100
epochs=20  
# x_train.shape[0]//batch_size
# Model Training 
history_pretrained = model_pretrained.fit(train_generator.flow(x_train, y_train, batch_size=batch_size),
                    epochs=epochs,
                    steps_per_epoch=len(x_train)//batch_size, 
                    validation_data=val_generator.flow(x_val, y_val, batch_size=batch_size),
                    validation_steps=len(x_val)//batch_size, 
                    callbacks=[lr_reduce], verbose=1 
                    )
```

### 11. Accuracy and Loss Graph

```python
# Accuracy and Loss graph For Model_1 
accuracy = history_pretrained.history['accuracy']
val_accuracy = history_pretrained.history['val_accuracy']

loss = history_pretrained.history['loss']
val_loss = history_pretrained.history['val_loss']

epochs = range(len(accuracy))
plt.plot(epochs, accuracy, 'go', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'g', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'go', label='Training loss')
plt.plot(epochs, val_loss, 'g', label='validation loss')
plt.title('training and validation loss')
plt.legend()
plt.show()
```

### 12. Evaluate with Test data

```python
# For model_1 
model_test_eval = model_pretrained.evaluate(x_test, y_test, verbose=1)
print("Test loss is :{}".format(model_test_eval[0])) 
print("Test accuracy is :{}".format(model_test_eval[1]))  
```
